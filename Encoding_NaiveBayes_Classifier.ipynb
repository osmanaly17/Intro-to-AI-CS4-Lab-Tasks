```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder


data = {'Weather': ['Sunny', 'Rainy', 'Snowy', 'Sunny', 'Rainy', 'Snowy'],
        'Target': ['Class1', 'Class2', 'Class1', 'Class2', 'Class1', 'Class2']}
df = pd.DataFrame(data)

print(df)


le = LabelEncoder()
df['Weather_LabelEncoded'] = le.fit_transform(df['Weather'])

print("\nLabel Encoded Data:")
print(df)


ohe = OneHotEncoder(sparse_output=False)
weather_ohe = ohe.fit_transform(df[['Weather']])


weather_ohe_df = pd.DataFrame(weather_ohe, columns=ohe.get_feature_names_out(['Weather']))
print("\nOne-Hot Encoded Data:")
print(weather_ohe_df)
```

```python
from sklearn.preprocessing import OrdinalEncoder


data = {'Priority': ['Low', 'Medium', 'High', 'Low', 'Medium', 'High'],
        'Target': ['Class1', 'Class2', 'Class1', 'Class2', 'Class1', 'Class2']}
df2 = pd.DataFrame(data)

print(df2)


ordinal_encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])
df2['Priority_Encoded'] = ordinal_encoder.fit_transform(df2[['Priority']])

print("\nOrdinal Encoded Data:")
print(df2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df[['Weather_LabelEncoded']]
y = df['Target']


le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)


X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)


model = GaussianNB()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


print("\nAccuracy:", accuracy_score(y_test, y_pred))


decoded_preds = le_target.inverse_transform(y_pred)
print("\nDecoded Predictions:", decoded_preds)
```
